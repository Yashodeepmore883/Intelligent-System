```python
class AstarGraph(object):
    def __init__(self):
        self.barriers = []
        t = int(input("Enter Total number of barriers: "))
        for i in range(t):
            x = int(input("Enter row number for barrier points: "))
            y = int(input("Enter column number for barrier points: "))
            self.barriers.append((x, y))  # Append as a tuple

    def heuristic(self, start, goal):
        D = 1
        dx = abs(start[0] - goal[0])
        dy = abs(start[1] - goal[1])
        return D * (dx + dy)

    def get_vertex_neighbours(self, pos):
        n = []
        for dx, dy in [(1, 0), (-1, 0), (0, 1), (0, -1)]:
            x2 = pos[0] + dx
            y2 = pos[1] + dy
            if 0 <= x2 <= 5 and 0 <= y2 <= 5 and (x2, y2) not in self.barriers:
                n.append((x2, y2))
        return n

    def move_cost(self, a, b):
        return 1

    def AStarSearch(self, start, end):
        G = {}
        F = {}

        G[start] = 0
        F[start] = self.heuristic(start, end)
        closedVertices = set()
        openVertices = set([start])
        cameFrom = {}

        while len(openVertices) > 0:
            current = None
            currentFscore = None
            for pos in openVertices:
                if current is None or F[pos] < currentFscore:
                    currentFscore = F[pos]
                    current = pos
            if current == end:
                path = [current]
                while current in cameFrom:
                    current = cameFrom[current]
                    path.append(current)
                path.reverse()
                return path, F[end]
            openVertices.remove(current)
            closedVertices.add(current)

            for neighbour in self.get_vertex_neighbours(current):
                if neighbour in closedVertices:
                    continue
                candidateG = G[current] + self.move_cost(current, neighbour)

                if neighbour not in openVertices:
                    openVertices.add(neighbour)
                elif candidateG >= G[neighbour]:
                    continue
                cameFrom[neighbour] = current
                G[neighbour] = candidateG
                H = self.heuristic(neighbour, end)
                F[neighbour] = G[neighbour] + H
        raise RuntimeError("A* failed to find a solution")

if __name__ == "__main__":
    graph = AstarGraph()
    s1 = int(input("Enter row number for start point: "))
    s2 = int(input("Enter column number for start point: "))
    g1 = int(input("Enter row number for goal point: "))
    g2 = int(input("Enter column number for goal point: "))
    result, cost = graph.AStarSearch((s1, s2), (g1, g2))
    print("route:", result)
    print("cost:", cost)
```
